```python
# Trio 1 — code cell: R1 (2021 regime dummy) re-estimation per spec §7 R1 + §6 v1.0.2 κ-tightening
#
# Estimator R1 (spec §6 R1 + §7 R1):
#   Y_p_logit = α + β_6·X_lag6 + β_9·X_lag9 + β_12·X_lag12 + β_regime·regime_2021 + ε
#   regime_2021 = 1{t ≥ 2021-01}; HAC bandwidth L = 12 per spec §3.4
#
# Composite (spec §3.5):  β_composite_R1 = β_6 + β_9 + β_12 (excludes regime dummy)
#                          c_R1 = (0, 1, 1, 1, 0)'  (zero loading on intercept and regime)
#
# Sign-AGREE comparison per spec §7.1 + §6 v1.0.2 κ-tightening:
#   r1_agreement = (sign(β_composite_R1) == sign(β_composite_primary))
#   If FALSE  → §3.3 Clause-A FIRES regardless of §7.1 aggregate classification.
#   If TRUE   → necessary-not-sufficient (R3 must also AGREE for full κ-clearance per §6 v1.0.2).

from __future__ import annotations

import numpy as np
import pandas as pd
import scipy.stats as stats
import statsmodels.api as sm
import matplotlib.pyplot as plt

from env import (
    verify_panel,
    PAIR_D_BETA_COMPOSITE,
    PAIR_D_HAC_SE,
    PAIR_D_T_STAT,
    PAIR_D_P_ONE,
    N_EXPECTED_POST_LAG_12,
)

# ── §2.1. Panel re-load via verify_panel() (re-asserts N=134 + zero nulls) ──
panel = verify_panel()
assert len(panel) == N_EXPECTED_POST_LAG_12, (
    f"panel row count {len(panel)} != expected {N_EXPECTED_POST_LAG_12} (spec §3.6)"
)
print(f"Panel loaded: N={len(panel)} rows; sample window "
      f"{panel['year_month'].min().strftime('%Y-%m')} → "
      f"{panel['year_month'].max().strftime('%Y-%m')}")

# ── §2.2. Construct regime_2021 dummy from REAL year_month values ───────────
# regime_2021 = 1 for t ≥ 2021-01-31 (Marco-2018 era under DANE methodology break per spec §6)
regime_2021 = (
    pd.to_datetime(panel["year_month"]) >= pd.Timestamp("2021-01-31", tz="UTC")
).astype(int).to_numpy()

# Verify segment sizes per spec §6 (2015-01 → 2020-12 = 72 months pre, 2021-01 → 2026-03 = 63 months post)
n_pre_2021 = int((regime_2021 == 0).sum())
n_post_2021 = int((regime_2021 == 1).sum())
print(f"\nRegime-2021 dummy segments: n_pre_2021={n_pre_2021}, n_post_2021={n_post_2021}, "
      f"total={n_pre_2021 + n_post_2021}")
print(f"  (expected: pre ≈ 72 months [2015-01 → 2020-12], post ≈ 62 months [2021-01 → 2026-03 - lag drop])")

# Confirm post-2021 segment captures the 94-cell rare-event month per NB01 Trio 2 acknowledgment
post_dates = panel.loc[regime_2021 == 1, "year_month"]
rare_event_in_post = bool((pd.to_datetime(post_dates) == pd.Timestamp("2024-10-31", tz="UTC")).any())
print(f"  94-cell rare-event month (2024-10-31) in post-2021 segment? {rare_event_in_post}  "
      f"(R1 captures methodology-break × rare-event interaction per spec §9.17(c))")

# ── §2.3. Re-fit PRIMARY baseline (no regime dummy) to ground sign-AGREE check ─
# This recomputes the primary OLS in NB03 to ensure deterministic comparison; matches
# Pair D R-trio precedent (β_composite_primary = β_6 + β_9 + β_12, HAC L=12).
y = panel["Y_p_logit"].to_numpy(dtype=float)
X_primary_no_const = panel[["X_lag6", "X_lag9", "X_lag12"]].to_numpy(dtype=float)
X_primary = sm.add_constant(X_primary_no_const, has_constant="add")  # (134, 4)

result_primary = sm.OLS(y, X_primary).fit(cov_type="HAC", cov_kwds={"maxlags": 12})
c_primary = np.array([0.0, 1.0, 1.0, 1.0])
beta_composite_primary = float(c_primary @ result_primary.params)
cov_primary = np.asarray(result_primary.cov_params())
se_composite_primary = float(np.sqrt(c_primary @ cov_primary @ c_primary))
t_composite_primary = beta_composite_primary / se_composite_primary
p_one_primary = float(1.0 - stats.norm.cdf(t_composite_primary))

print("\n=== PRIMARY baseline (re-fit in NB03 for sign-AGREE grounding) ===")
print(f"  β_composite_primary    = {beta_composite_primary:+.8f}")
print(f"  HAC SE_composite_primary = {se_composite_primary:.8f}")
print(f"  t_composite_primary    = {t_composite_primary:+.6f}")
print(f"  p_one_HAC_primary      = {p_one_primary:.6e}")
print(f"  sign(β_composite_primary) = {'+' if beta_composite_primary > 0 else '-' if beta_composite_primary < 0 else '0'}")

# ── §2.4. R1 design matrix: (intercept, X_lag6, X_lag9, X_lag12, regime_2021) ─
# R1 specification per spec §6 + §7: add regime_2021 as additional regressor
# on TOP of primary (X_lag6, X_lag9, X_lag12) design (Path B per Task 2.4 Step 1 dispatch).
X_R1_no_const = np.column_stack([
    panel[["X_lag6", "X_lag9", "X_lag12"]].to_numpy(dtype=float),
    regime_2021.astype(float),
])
X_R1 = sm.add_constant(X_R1_no_const, has_constant="add")  # (134, 5)
assert X_R1.shape == (134, 5), f"R1 design matrix shape {X_R1.shape} != (134, 5)"
print(f"\nR1 design matrix: shape={X_R1.shape} (columns: const, X_lag6, X_lag9, X_lag12, regime_2021)")

# ── §2.5. R1 OLS fit with HAC SE (Newey-West, L=12 per spec §3.4) ───────────
result_R1 = sm.OLS(y, X_R1).fit(cov_type="HAC", cov_kwds={"maxlags": 12})
alpha_R1 = float(result_R1.params[0])
beta_6_R1 = float(result_R1.params[1])
beta_9_R1 = float(result_R1.params[2])
beta_12_R1 = float(result_R1.params[3])
beta_regime_R1 = float(result_R1.params[4])
se_alpha_R1 = float(result_R1.bse[0])
se_beta_6_R1 = float(result_R1.bse[1])
se_beta_9_R1 = float(result_R1.bse[2])
se_beta_12_R1 = float(result_R1.bse[3])
se_beta_regime_R1 = float(result_R1.bse[4])

print("\n=== R1 individual coefficients (HAC SE, L=12) ===")
print(f"  α_R1 (intercept) = {alpha_R1:+.8f}  (HAC SE {se_alpha_R1:.6f}; "
      f"t={alpha_R1/se_alpha_R1:+.4f})")
print(f"  β_6_R1           = {beta_6_R1:+.8f}  (HAC SE {se_beta_6_R1:.6f}; "
      f"t={beta_6_R1/se_beta_6_R1:+.4f})")
print(f"  β_9_R1           = {beta_9_R1:+.8f}  (HAC SE {se_beta_9_R1:.6f}; "
      f"t={beta_9_R1/se_beta_9_R1:+.4f})")
print(f"  β_12_R1          = {beta_12_R1:+.8f}  (HAC SE {se_beta_12_R1:.6f}; "
      f"t={beta_12_R1/se_beta_12_R1:+.4f})")
print(f"  β_regime_R1      = {beta_regime_R1:+.8f}  (HAC SE {se_beta_regime_R1:.6f}; "
      f"t={beta_regime_R1/se_beta_regime_R1:+.4f})  [INFORMATIONAL — residual empalme-bias diagnostic]")

# ── §2.6. R1 composite β + composite HAC SE via linear restriction ───────────
# c_R1 = (0, 1, 1, 1, 0)' — zero loading on intercept and regime; unit loading on β_6/9/12.
c_R1 = np.array([0.0, 1.0, 1.0, 1.0, 0.0])
beta_composite_R1 = float(c_R1 @ result_R1.params)
cov_R1 = np.asarray(result_R1.cov_params())
var_composite_R1 = float(c_R1 @ cov_R1 @ c_R1)
assert var_composite_R1 > 0, f"R1 composite variance non-positive: {var_composite_R1}"
se_composite_R1 = float(np.sqrt(var_composite_R1))
t_composite_R1 = beta_composite_R1 / se_composite_R1
p_one_R1 = float(1.0 - stats.norm.cdf(t_composite_R1))

print("\n=== R1 composite β (spec §3.5: β_composite_R1 = β_6_R1 + β_9_R1 + β_12_R1) ===")
print(f"  β_composite_R1         = {beta_composite_R1:+.8f}")
print(f"  HAC SE_composite_R1    = {se_composite_R1:.8f}")
print(f"  t_composite_R1         = {t_composite_R1:+.6f}")
print(f"  p_one_HAC_R1           = {p_one_R1:.6e}")
print(f"  sign(β_composite_R1)   = {'+' if beta_composite_R1 > 0 else '-' if beta_composite_R1 < 0 else '0'}")

# ── §2.7. Sign-AGREE comparison primary-vs-R1 per spec §7.1 + §6 v1.0.2 κ-tightening ─
sign_primary = int(np.sign(beta_composite_primary))
sign_R1 = int(np.sign(beta_composite_R1))
r1_agreement = bool(sign_primary == sign_R1)

print("\n=== Sign-AGREE comparison primary-vs-R1 (spec §7.1 + §6 v1.0.2 κ-tightening) ===")
print(f"  sign(β_composite_primary) = {sign_primary:+d}  ({beta_composite_primary:+.8f})")
print(f"  sign(β_composite_R1)      = {sign_R1:+d}  ({beta_composite_R1:+.8f})")
print(f"  R1 sign-AGREEs with primary? {r1_agreement}")

# ── §2.8. §3.3 Clause-A FIRES status under v1.0.2 κ-tightening ──────────────
if not r1_agreement:
    clause_a_status = (
        "Clause-A WILL FIRE per §6 v1.0.2 κ-tightening (R1 sign-DIFFERENT from primary; "
        "MANDATORY ESCALATE regardless of §7.1 aggregate classification; no judgment-call escape)."
    )
else:
    clause_a_status = (
        "Clause-A NOT triggered by R1 alone; R1 sign-AGREES with primary. "
        "Final κ-clearance pending R3 sign-AGREE per §6 v1.0.2 (R1 sign-AGREE is necessary-not-sufficient)."
    )
print(f"\n=== §3.3 Clause-A status (CORRECTIONS-κ v1.0.2) ===")
print(f"  {clause_a_status}")

# ── §2.9. β_regime_R1 interpretation (residual empalme-bias diagnostic per spec §6) ─
print("\n=== β_regime_R1 informational interpretation (spec §6 residual-bias acknowledgment) ===")
if abs(beta_regime_R1 / se_beta_regime_R1) > 1.96:
    sig_label = f"statistically significant at 95% (|t|={abs(beta_regime_R1/se_beta_regime_R1):.2f} > 1.96)"
else:
    sig_label = f"not significant at 95% (|t|={abs(beta_regime_R1/se_beta_regime_R1):.2f} ≤ 1.96)"
direction = "post-2021 era HIGHER Y_p_logit conditional on X-lags" if beta_regime_R1 > 0 else "post-2021 era LOWER Y_p_logit conditional on X-lags"
print(f"  β_regime_R1 = {beta_regime_R1:+.6f}; {sig_label}")
print(f"  Interpretation: {direction} (residual regime-shift not fully captured by empalme).")

# ── §2.10. Pair D contrast (R-AGREE 0/4 baseline per memory project_pair_d_phase2_pass) ─
print("\n=== Pair D R-AGREE 0/4 sign-flips contrast (memory project_pair_d_phase2_pass) ===")
print(f"  Pair D primary β_composite     = {PAIR_D_BETA_COMPOSITE:+.8f}  "
      f"(HAC SE {PAIR_D_HAC_SE:.5f}; t={PAIR_D_T_STAT:+.4f}; p_one={PAIR_D_P_ONE:.2e})")
print(f"  Pair D R-AGREE 0/4 sign-flips: R1 sign-AGREEd with primary at PASS verdict.")
print(f"  This iteration R1 vs primary: {'AGREES' if r1_agreement else 'DISAGREES'} "
      f"(Section J narrowing may produce different R1 outcome).")

# ── §2.11. Plot 1 — bar chart of R1 vs primary individual β_k for k ∈ {6, 9, 12} ─
labels = ["β_6 (lag 6)", "β_9 (lag 9)", "β_12 (lag 12)"]
primary_vals = [
    float(result_primary.params[1]),
    float(result_primary.params[2]),
    float(result_primary.params[3]),
]
primary_errs = [
    1.96 * float(result_primary.bse[1]),
    1.96 * float(result_primary.bse[2]),
    1.96 * float(result_primary.bse[3]),
]
r1_vals = [beta_6_R1, beta_9_R1, beta_12_R1]
r1_errs = [1.96 * se_beta_6_R1, 1.96 * se_beta_9_R1, 1.96 * se_beta_12_R1]

x_pos = np.arange(len(labels))
bar_w = 0.38
fig, ax = plt.subplots(figsize=(9.0, 4.6))
ax.bar(x_pos - bar_w / 2, primary_vals, bar_w, yerr=primary_errs, capsize=5,
       color="#1f4e79", alpha=0.85, edgecolor="#202020", linewidth=0.8, label="Primary (no regime)")
ax.bar(x_pos + bar_w / 2, r1_vals, bar_w, yerr=r1_errs, capsize=5,
       color="#a36b00", alpha=0.85, edgecolor="#202020", linewidth=0.8, label="R1 (+ regime_2021)")
ax.axhline(0, color="#404040", linewidth=0.6)
ax.set_xticks(x_pos)
ax.set_xticklabels(labels)
ax.set_ylabel("β̂ (logit-Y per unit log-COP/USD)")
ax.set_title(
    f"Per-lag β̂ comparison: primary vs R1 (95% HAC CI error bars)\n"
    f"primary β_composite = {beta_composite_primary:+.4f}; "
    f"R1 β_composite_R1 = {beta_composite_R1:+.4f}; "
    f"R1 sign-AGREE = {r1_agreement}"
)
ax.legend(loc="best", fontsize=9)
ax.grid(True, alpha=0.3, axis="y")
plt.tight_layout()
plt.show()

# ── §2.12. Plot 2 — scatter of R1 composite β vs primary composite β with y=x line ─
# (single-point scatter; informational visualization of the R1 vs primary delta)
fig, ax = plt.subplots(figsize=(6.0, 6.0))
xr = np.array([
    min(beta_composite_primary, beta_composite_R1) - 0.05 * abs(beta_composite_primary + beta_composite_R1 + 1e-6),
    max(beta_composite_primary, beta_composite_R1) + 0.05 * abs(beta_composite_primary + beta_composite_R1 + 1e-6),
])
ax.plot(xr, xr, color="#404040", linestyle="--", linewidth=1.0, label="y = x (perfect agreement)")
ax.errorbar(
    [beta_composite_primary], [beta_composite_R1],
    xerr=[1.96 * se_composite_primary], yerr=[1.96 * se_composite_R1],
    fmt="o", color="#c00000", markersize=10, capsize=6, ecolor="#c00000", linewidth=1.5,
    label=f"(primary, R1) = ({beta_composite_primary:+.4f}, {beta_composite_R1:+.4f})",
)
ax.axhline(0, color="#404040", linewidth=0.4, alpha=0.5)
ax.axvline(0, color="#404040", linewidth=0.4, alpha=0.5)
ax.set_xlabel("β_composite_primary")
ax.set_ylabel("β_composite_R1")
ax.set_title(
    f"R1 vs primary composite-β (95% HAC CIs)\n"
    f"sign-AGREE? {r1_agreement} | Pair D R-AGREE 0/4 baseline (project_pair_d_phase2_pass)"
)
ax.legend(loc="best", fontsize=9)
ax.grid(True, alpha=0.3)
ax.set_aspect("equal", adjustable="datalim")
plt.tight_layout()
plt.show()

# ── §2.13. Trio 1 readout summary (handoff to Trio 2 R2) ─────────────────────
print("\n=== Trio 1 readout summary ===")
print(f"  R1 composite β   : {beta_composite_R1:+.8f}  (HAC SE {se_composite_R1:.6f}; "
      f"t={t_composite_R1:+.4f}; p_one={p_one_R1:.4e})")
print(f"  R1 sign-AGREE    : {r1_agreement}  (primary sign={sign_primary:+d}, R1 sign={sign_R1:+d})")
print(f"  β_regime_R1      : {beta_regime_R1:+.6f}  (HAC SE {se_beta_regime_R1:.6f}; "
      f"informational residual-empalme-bias diagnostic)")
print(f"  Clause-A status  : {'WILL FIRE (R1 disagrees)' if not r1_agreement else 'NOT triggered by R1; pending R3'}")
print(f"  R-row count toward §7.1 AGREE classification: {1 if r1_agreement else 0}/4 so far "
      f"(threshold ≥3/4 sign-preserving for AGREE).")
```
