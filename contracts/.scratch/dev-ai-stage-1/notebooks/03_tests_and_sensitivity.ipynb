```python
# Trio 1 — code cell: R1 (2021 regime dummy) re-estimation per spec §7 R1 + §6 v1.0.2 κ-tightening
#
# Estimator R1 (spec §6 R1 + §7 R1):
#   Y_p_logit = α + β_6·X_lag6 + β_9·X_lag9 + β_12·X_lag12 + β_regime·regime_2021 + ε
#   regime_2021 = 1{t ≥ 2021-01}; HAC bandwidth L = 12 per spec §3.4
#
# Composite (spec §3.5):  β_composite_R1 = β_6 + β_9 + β_12 (excludes regime dummy)
#                          c_R1 = (0, 1, 1, 1, 0)'  (zero loading on intercept and regime)
#
# Sign-AGREE comparison per spec §7.1 + §6 v1.0.2 κ-tightening:
#   r1_agreement = (sign(β_composite_R1) == sign(β_composite_primary))
#   If FALSE  → §3.3 Clause-A FIRES regardless of §7.1 aggregate classification.
#   If TRUE   → necessary-not-sufficient (R3 must also AGREE for full κ-clearance per §6 v1.0.2).

from __future__ import annotations

import numpy as np
import pandas as pd
import scipy.stats as stats
import statsmodels.api as sm
import matplotlib.pyplot as plt

from env import (
    verify_panel,
    PAIR_D_BETA_COMPOSITE,
    PAIR_D_HAC_SE,
    PAIR_D_T_STAT,
    PAIR_D_P_ONE,
    N_EXPECTED_POST_LAG_12,
)

# ── §2.1. Panel re-load via verify_panel() (re-asserts N=134 + zero nulls) ──
panel = verify_panel()
assert len(panel) == N_EXPECTED_POST_LAG_12, (
    f"panel row count {len(panel)} != expected {N_EXPECTED_POST_LAG_12} (spec §3.6)"
)
print(f"Panel loaded: N={len(panel)} rows; sample window "
      f"{panel['year_month'].min().strftime('%Y-%m')} → "
      f"{panel['year_month'].max().strftime('%Y-%m')}")

# ── §2.2. Construct regime_2021 dummy from REAL year_month values ───────────
# regime_2021 = 1 for t ≥ 2021-01-31 (Marco-2018 era under DANE methodology break per spec §6)
regime_2021 = (
    pd.to_datetime(panel["year_month"]) >= pd.Timestamp("2021-01-31", tz="UTC")
).astype(int).to_numpy()

# Verify segment sizes per spec §6 (2015-01 → 2020-12 = 72 months pre, 2021-01 → 2026-03 = 63 months post)
n_pre_2021 = int((regime_2021 == 0).sum())
n_post_2021 = int((regime_2021 == 1).sum())
print(f"\nRegime-2021 dummy segments: n_pre_2021={n_pre_2021}, n_post_2021={n_post_2021}, "
      f"total={n_pre_2021 + n_post_2021}")
print(f"  (expected: pre ≈ 72 months [2015-01 → 2020-12], post ≈ 62 months [2021-01 → 2026-03 - lag drop])")

# Confirm post-2021 segment captures the 94-cell rare-event month per NB01 Trio 2 acknowledgment
post_dates = panel.loc[regime_2021 == 1, "year_month"]
rare_event_in_post = bool((pd.to_datetime(post_dates) == pd.Timestamp("2024-10-31", tz="UTC")).any())
print(f"  94-cell rare-event month (2024-10-31) in post-2021 segment? {rare_event_in_post}  "
      f"(R1 captures methodology-break × rare-event interaction per spec §9.17(c))")

# ── §2.3. Re-fit PRIMARY baseline (no regime dummy) to ground sign-AGREE check ─
# This recomputes the primary OLS in NB03 to ensure deterministic comparison; matches
# Pair D R-trio precedent (β_composite_primary = β_6 + β_9 + β_12, HAC L=12).
y = panel["Y_p_logit"].to_numpy(dtype=float)
X_primary_no_const = panel[["X_lag6", "X_lag9", "X_lag12"]].to_numpy(dtype=float)
X_primary = sm.add_constant(X_primary_no_const, has_constant="add")  # (134, 4)

result_primary = sm.OLS(y, X_primary).fit(cov_type="HAC", cov_kwds={"maxlags": 12})
c_primary = np.array([0.0, 1.0, 1.0, 1.0])
beta_composite_primary = float(c_primary @ result_primary.params)
cov_primary = np.asarray(result_primary.cov_params())
se_composite_primary = float(np.sqrt(c_primary @ cov_primary @ c_primary))
t_composite_primary = beta_composite_primary / se_composite_primary
p_one_primary = float(1.0 - stats.norm.cdf(t_composite_primary))

print("\n=== PRIMARY baseline (re-fit in NB03 for sign-AGREE grounding) ===")
print(f"  β_composite_primary    = {beta_composite_primary:+.8f}")
print(f"  HAC SE_composite_primary = {se_composite_primary:.8f}")
print(f"  t_composite_primary    = {t_composite_primary:+.6f}")
print(f"  p_one_HAC_primary      = {p_one_primary:.6e}")
print(f"  sign(β_composite_primary) = {'+' if beta_composite_primary > 0 else '-' if beta_composite_primary < 0 else '0'}")

# ── §2.4. R1 design matrix: (intercept, X_lag6, X_lag9, X_lag12, regime_2021) ─
# R1 specification per spec §6 + §7: add regime_2021 as additional regressor
# on TOP of primary (X_lag6, X_lag9, X_lag12) design (Path B per Task 2.4 Step 1 dispatch).
X_R1_no_const = np.column_stack([
    panel[["X_lag6", "X_lag9", "X_lag12"]].to_numpy(dtype=float),
    regime_2021.astype(float),
])
X_R1 = sm.add_constant(X_R1_no_const, has_constant="add")  # (134, 5)
assert X_R1.shape == (134, 5), f"R1 design matrix shape {X_R1.shape} != (134, 5)"
print(f"\nR1 design matrix: shape={X_R1.shape} (columns: const, X_lag6, X_lag9, X_lag12, regime_2021)")

# ── §2.5. R1 OLS fit with HAC SE (Newey-West, L=12 per spec §3.4) ───────────
result_R1 = sm.OLS(y, X_R1).fit(cov_type="HAC", cov_kwds={"maxlags": 12})
alpha_R1 = float(result_R1.params[0])
beta_6_R1 = float(result_R1.params[1])
beta_9_R1 = float(result_R1.params[2])
beta_12_R1 = float(result_R1.params[3])
beta_regime_R1 = float(result_R1.params[4])
se_alpha_R1 = float(result_R1.bse[0])
se_beta_6_R1 = float(result_R1.bse[1])
se_beta_9_R1 = float(result_R1.bse[2])
se_beta_12_R1 = float(result_R1.bse[3])
se_beta_regime_R1 = float(result_R1.bse[4])

print("\n=== R1 individual coefficients (HAC SE, L=12) ===")
print(f"  α_R1 (intercept) = {alpha_R1:+.8f}  (HAC SE {se_alpha_R1:.6f}; "
      f"t={alpha_R1/se_alpha_R1:+.4f})")
print(f"  β_6_R1           = {beta_6_R1:+.8f}  (HAC SE {se_beta_6_R1:.6f}; "
      f"t={beta_6_R1/se_beta_6_R1:+.4f})")
print(f"  β_9_R1           = {beta_9_R1:+.8f}  (HAC SE {se_beta_9_R1:.6f}; "
      f"t={beta_9_R1/se_beta_9_R1:+.4f})")
print(f"  β_12_R1          = {beta_12_R1:+.8f}  (HAC SE {se_beta_12_R1:.6f}; "
      f"t={beta_12_R1/se_beta_12_R1:+.4f})")
print(f"  β_regime_R1      = {beta_regime_R1:+.8f}  (HAC SE {se_beta_regime_R1:.6f}; "
      f"t={beta_regime_R1/se_beta_regime_R1:+.4f})  [INFORMATIONAL — residual empalme-bias diagnostic]")

# ── §2.6. R1 composite β + composite HAC SE via linear restriction ───────────
# c_R1 = (0, 1, 1, 1, 0)' — zero loading on intercept and regime; unit loading on β_6/9/12.
c_R1 = np.array([0.0, 1.0, 1.0, 1.0, 0.0])
beta_composite_R1 = float(c_R1 @ result_R1.params)
cov_R1 = np.asarray(result_R1.cov_params())
var_composite_R1 = float(c_R1 @ cov_R1 @ c_R1)
assert var_composite_R1 > 0, f"R1 composite variance non-positive: {var_composite_R1}"
se_composite_R1 = float(np.sqrt(var_composite_R1))
t_composite_R1 = beta_composite_R1 / se_composite_R1
p_one_R1 = float(1.0 - stats.norm.cdf(t_composite_R1))

print("\n=== R1 composite β (spec §3.5: β_composite_R1 = β_6_R1 + β_9_R1 + β_12_R1) ===")
print(f"  β_composite_R1         = {beta_composite_R1:+.8f}")
print(f"  HAC SE_composite_R1    = {se_composite_R1:.8f}")
print(f"  t_composite_R1         = {t_composite_R1:+.6f}")
print(f"  p_one_HAC_R1           = {p_one_R1:.6e}")
print(f"  sign(β_composite_R1)   = {'+' if beta_composite_R1 > 0 else '-' if beta_composite_R1 < 0 else '0'}")

# ── §2.7. Sign-AGREE comparison primary-vs-R1 per spec §7.1 + §6 v1.0.2 κ-tightening ─
sign_primary = int(np.sign(beta_composite_primary))
sign_R1 = int(np.sign(beta_composite_R1))
r1_agreement = bool(sign_primary == sign_R1)

print("\n=== Sign-AGREE comparison primary-vs-R1 (spec §7.1 + §6 v1.0.2 κ-tightening) ===")
print(f"  sign(β_composite_primary) = {sign_primary:+d}  ({beta_composite_primary:+.8f})")
print(f"  sign(β_composite_R1)      = {sign_R1:+d}  ({beta_composite_R1:+.8f})")
print(f"  R1 sign-AGREEs with primary? {r1_agreement}")

# ── §2.8. §3.3 Clause-A FIRES status under v1.0.2 κ-tightening ──────────────
if not r1_agreement:
    clause_a_status = (
        "Clause-A WILL FIRE per §6 v1.0.2 κ-tightening (R1 sign-DIFFERENT from primary; "
        "MANDATORY ESCALATE regardless of §7.1 aggregate classification; no judgment-call escape)."
    )
else:
    clause_a_status = (
        "Clause-A NOT triggered by R1 alone; R1 sign-AGREES with primary. "
        "Final κ-clearance pending R3 sign-AGREE per §6 v1.0.2 (R1 sign-AGREE is necessary-not-sufficient)."
    )
print(f"\n=== §3.3 Clause-A status (CORRECTIONS-κ v1.0.2) ===")
print(f"  {clause_a_status}")

# ── §2.9. β_regime_R1 interpretation (residual empalme-bias diagnostic per spec §6) ─
print("\n=== β_regime_R1 informational interpretation (spec §6 residual-bias acknowledgment) ===")
if abs(beta_regime_R1 / se_beta_regime_R1) > 1.96:
    sig_label = f"statistically significant at 95% (|t|={abs(beta_regime_R1/se_beta_regime_R1):.2f} > 1.96)"
else:
    sig_label = f"not significant at 95% (|t|={abs(beta_regime_R1/se_beta_regime_R1):.2f} ≤ 1.96)"
direction = "post-2021 era HIGHER Y_p_logit conditional on X-lags" if beta_regime_R1 > 0 else "post-2021 era LOWER Y_p_logit conditional on X-lags"
print(f"  β_regime_R1 = {beta_regime_R1:+.6f}; {sig_label}")
print(f"  Interpretation: {direction} (residual regime-shift not fully captured by empalme).")

# ── §2.10. Pair D contrast (R-AGREE 0/4 baseline per memory project_pair_d_phase2_pass) ─
print("\n=== Pair D R-AGREE 0/4 sign-flips contrast (memory project_pair_d_phase2_pass) ===")
print(f"  Pair D primary β_composite     = {PAIR_D_BETA_COMPOSITE:+.8f}  "
      f"(HAC SE {PAIR_D_HAC_SE:.5f}; t={PAIR_D_T_STAT:+.4f}; p_one={PAIR_D_P_ONE:.2e})")
print(f"  Pair D R-AGREE 0/4 sign-flips: R1 sign-AGREEd with primary at PASS verdict.")
print(f"  This iteration R1 vs primary: {'AGREES' if r1_agreement else 'DISAGREES'} "
      f"(Section J narrowing may produce different R1 outcome).")

# ── §2.11. Plot 1 — bar chart of R1 vs primary individual β_k for k ∈ {6, 9, 12} ─
labels = ["β_6 (lag 6)", "β_9 (lag 9)", "β_12 (lag 12)"]
primary_vals = [
    float(result_primary.params[1]),
    float(result_primary.params[2]),
    float(result_primary.params[3]),
]
primary_errs = [
    1.96 * float(result_primary.bse[1]),
    1.96 * float(result_primary.bse[2]),
    1.96 * float(result_primary.bse[3]),
]
r1_vals = [beta_6_R1, beta_9_R1, beta_12_R1]
r1_errs = [1.96 * se_beta_6_R1, 1.96 * se_beta_9_R1, 1.96 * se_beta_12_R1]

x_pos = np.arange(len(labels))
bar_w = 0.38
fig, ax = plt.subplots(figsize=(9.0, 4.6))
ax.bar(x_pos - bar_w / 2, primary_vals, bar_w, yerr=primary_errs, capsize=5,
       color="#1f4e79", alpha=0.85, edgecolor="#202020", linewidth=0.8, label="Primary (no regime)")
ax.bar(x_pos + bar_w / 2, r1_vals, bar_w, yerr=r1_errs, capsize=5,
       color="#a36b00", alpha=0.85, edgecolor="#202020", linewidth=0.8, label="R1 (+ regime_2021)")
ax.axhline(0, color="#404040", linewidth=0.6)
ax.set_xticks(x_pos)
ax.set_xticklabels(labels)
ax.set_ylabel("β̂ (logit-Y per unit log-COP/USD)")
ax.set_title(
    f"Per-lag β̂ comparison: primary vs R1 (95% HAC CI error bars)\n"
    f"primary β_composite = {beta_composite_primary:+.4f}; "
    f"R1 β_composite_R1 = {beta_composite_R1:+.4f}; "
    f"R1 sign-AGREE = {r1_agreement}"
)
ax.legend(loc="best", fontsize=9)
ax.grid(True, alpha=0.3, axis="y")
plt.tight_layout()
plt.show()

# ── §2.12. Plot 2 — scatter of R1 composite β vs primary composite β with y=x line ─
# (single-point scatter; informational visualization of the R1 vs primary delta)
fig, ax = plt.subplots(figsize=(6.0, 6.0))
xr = np.array([
    min(beta_composite_primary, beta_composite_R1) - 0.05 * abs(beta_composite_primary + beta_composite_R1 + 1e-6),
    max(beta_composite_primary, beta_composite_R1) + 0.05 * abs(beta_composite_primary + beta_composite_R1 + 1e-6),
])
ax.plot(xr, xr, color="#404040", linestyle="--", linewidth=1.0, label="y = x (perfect agreement)")
ax.errorbar(
    [beta_composite_primary], [beta_composite_R1],
    xerr=[1.96 * se_composite_primary], yerr=[1.96 * se_composite_R1],
    fmt="o", color="#c00000", markersize=10, capsize=6, ecolor="#c00000", linewidth=1.5,
    label=f"(primary, R1) = ({beta_composite_primary:+.4f}, {beta_composite_R1:+.4f})",
)
ax.axhline(0, color="#404040", linewidth=0.4, alpha=0.5)
ax.axvline(0, color="#404040", linewidth=0.4, alpha=0.5)
ax.set_xlabel("β_composite_primary")
ax.set_ylabel("β_composite_R1")
ax.set_title(
    f"R1 vs primary composite-β (95% HAC CIs)\n"
    f"sign-AGREE? {r1_agreement} | Pair D R-AGREE 0/4 baseline (project_pair_d_phase2_pass)"
)
ax.legend(loc="best", fontsize=9)
ax.grid(True, alpha=0.3)
ax.set_aspect("equal", adjustable="datalim")
plt.tight_layout()
plt.show()

# ── §2.13. Trio 1 readout summary (handoff to Trio 2 R2) ─────────────────────
print("\n=== Trio 1 readout summary ===")
print(f"  R1 composite β   : {beta_composite_R1:+.8f}  (HAC SE {se_composite_R1:.6f}; "
      f"t={t_composite_R1:+.4f}; p_one={p_one_R1:.4e})")
print(f"  R1 sign-AGREE    : {r1_agreement}  (primary sign={sign_primary:+d}, R1 sign={sign_R1:+d})")
print(f"  β_regime_R1      : {beta_regime_R1:+.6f}  (HAC SE {se_beta_regime_R1:.6f}; "
      f"informational residual-empalme-bias diagnostic)")
print(f"  Clause-A status  : {'WILL FIRE (R1 disagrees)' if not r1_agreement else 'NOT triggered by R1; pending R3'}")
print(f"  R-row count toward §7.1 AGREE classification: {1 if r1_agreement else 0}/4 so far "
      f"(threshold ≥3/4 sign-preserving for AGREE).")
```

```python
# Trio 6 — code cell: §5.5 escalation suite — D-i quantile τ=0.90 + D-ii GARCH(1,1)-X + D-iii EVT POT
#
# Spec refs: §5.5 (escalation methodology — D-i quantile τ=0.90; D-ii GARCH(1,1)-X composite mean-equation;
#                  D-iii EVT POT on primary-OLS residuals at empirical 0.90 residual quantile);
#            §3.4 (ESCALATE-PASS disjunctive threshold, all three disjuncts at one-sided p ≤ 0.10;
#                  structural-disjunction MHT defense — distinct distributional-moment parameters
#                  mapping to distinct convex-instrument designs);
#            §3.3 (verdict primitives: Clause-A and Clause-B trigger structure);
#            §9.6 (escalation as pre-authorization, NOT post-hoc rescue — runs regardless of mean-OLS routing);
#            §9.17 (Phase-3 result-memo §11.X disclosure pre-pin — escalation_results feeds disclosure block);
#            Plan v1.1.1 Task 2.4 Step 6 (emit-target output/ESCALATION_RESULTS.md).
#
# Self-grounding (per Trio 1-5 precedent): re-fit primary OLS deterministically in this cell;
# do NOT depend on prior trios' state being in scope. Cite β_composite_primary verbatim from
# the re-fit to anchor sign comparisons in the disjunction evaluation.

from __future__ import annotations

import datetime as _dt
import warnings
from pathlib import Path

import numpy as np
import pandas as pd
import scipy.stats as stats
import statsmodels.api as sm
from arch import arch_model
from scipy.stats import genpareto
from statsmodels.regression.quantile_regression import QuantReg

from env import (
    verify_panel,
    PAIR_D_BETA_COMPOSITE,
    N_EXPECTED_POST_LAG_12,
    OUTPUT_DIR,
)

# ── §7.1. Panel re-load via verify_panel() (re-asserts N=134 + zero nulls) ─
panel = verify_panel()
assert len(panel) == N_EXPECTED_POST_LAG_12, (
    f"panel row count {len(panel)} != expected {N_EXPECTED_POST_LAG_12} (spec §3.6)"
)
print(f"Panel loaded: N={len(panel)} rows; "
      f"window {panel['year_month'].min().strftime('%Y-%m')} → "
      f"{panel['year_month'].max().strftime('%Y-%m')}")

# ── §7.2. Re-fit PRIMARY (Y_p_logit on X_lag6/9/12) — sign anchor for §3.4 disjunction ──
y_p_logit = panel["Y_p_logit"].to_numpy(dtype=float)
X_no_const = panel[["X_lag6", "X_lag9", "X_lag12"]].to_numpy(dtype=float)
X_design = sm.add_constant(X_no_const, has_constant="add")  # (134, 4)

result_primary = sm.OLS(y_p_logit, X_design).fit(cov_type="HAC", cov_kwds={"maxlags": 12})
c_3lag = np.array([0.0, 1.0, 1.0, 1.0])
beta_composite_primary = float(c_3lag @ result_primary.params)
sign_primary = 1 if beta_composite_primary > 0 else -1 if beta_composite_primary < 0 else 0
print(f"\n=== Primary OLS sign anchor (§3.4 disjunction reference) ===")
print(f"  β_composite_primary  : {beta_composite_primary:+.8f}")
print(f"  sign_primary         : {'+' if sign_primary > 0 else '−' if sign_primary < 0 else '0'}")

# ── §7.3. D-i — Quantile regression τ=0.90 (spec §5.5 D-i + §3.4 D-i) ─────
# Per spec §3.4 D-i verbatim: "the τ = 0.90 quantile-regression coefficient on X_{t-9}
# (representative middle-of-window lag) is > 0 at one-sided p ≤ 0.10".
# Per spec §5.5 D-i verbatim: "conditional quantile regression of Y_t (logit) on the
# three lagged-X regressors, estimated at τ = 0.90".
# Operational reading: §5.5 fits the 3-lag joint specification; §3.4 evaluates the
# X_{t-9} (lag-9) coefficient as the spec-pinned "representative middle-of-window lag".
print(f"\n=== D-i — Quantile regression τ=0.90 ===")
qr_model = QuantReg(y_p_logit, X_design)
with warnings.catch_warnings():
    warnings.simplefilter("ignore")
    result_qr = qr_model.fit(q=0.90)

# Coefficient indices: [0]=const, [1]=X_lag6, [2]=X_lag9, [3]=X_lag12 (spec §3.4 D-i lag-9 anchor)
beta_qr_lag9 = float(result_qr.params[2])
se_qr_lag9 = float(result_qr.bse[2])
t_qr_lag9 = beta_qr_lag9 / se_qr_lag9 if se_qr_lag9 > 0 else float("nan")
# One-sided p (positive-direction alternative per §3.4 D-i): p_one = 1 - Φ(t)
# QuantReg result.pvalues is two-sided; convert correctly: if t > 0, p_one = pvalues/2; else 1 - pvalues/2.
p_two_qr_lag9 = float(result_qr.pvalues[2])
p_one_qr_lag9 = p_two_qr_lag9 / 2.0 if t_qr_lag9 > 0 else 1.0 - (p_two_qr_lag9 / 2.0)

# Composite τ=0.90 quantile β (informational — sum of three lag coefficients; primary §3.4 trigger is X_{t-9})
beta_qr_composite = float(c_3lag @ result_qr.params)
print(f"  β_qr_lag9 (τ=0.90)   : {beta_qr_lag9:+.8f}")
print(f"  SE_qr_lag9            : {se_qr_lag9:.8f}")
print(f"  t-stat_qr_lag9        : {t_qr_lag9:+.6f}")
print(f"  p_one_qr_lag9         : {p_one_qr_lag9:.8e}")
print(f"  β_qr_composite (info) : {beta_qr_composite:+.8f}")

sign_qr_lag9 = 1 if beta_qr_lag9 > 0 else -1 if beta_qr_lag9 < 0 else 0
d_i_sign_match = (sign_qr_lag9 == sign_primary) and (sign_qr_lag9 > 0)
d_i_pass = d_i_sign_match and (p_one_qr_lag9 <= 0.10)
print(f"  D-i sign match prim   : {d_i_sign_match} (sign_primary={sign_primary:+d}, sign_qr_lag9={sign_qr_lag9:+d})")
print(f"  D-i α=0.10 threshold  : {'PASS' if d_i_pass else 'FAIL'} (β>0 AND p_one≤0.10)")

# ── §7.4. D-ii — GARCH(1,1)-X (spec §5.5 D-ii + §3.4 D-ii) ─────────────────
# Per spec §3.4 D-ii verbatim: "in a GARCH(1,1) with log(COP_USD_{t-k}) for k ∈ {6, 9, 12}
# as exogenous mean-equation regressors, the composite GARCH-X mean β (sum of the three
# lag coefficients) is > 0 at one-sided p ≤ 0.10".
# Per spec §5.5 D-ii verbatim: "GARCH(1,1) on Y_t (logit) with the three lagged-X
# regressors entering the mean equation as exogenous covariates. Tests whether mean-β
# remains positive under explicit time-varying-volatility correction. The composite
# GARCH-X β is the sum of the three lag mean-equation coefficients."
print(f"\n=== D-ii — GARCH(1,1)-X composite mean β ===")
# arch.arch_model uses LS (constant) mean by default; pass exogenous regressors via x=
# Convergence is more robust on rescaled data; multiply Y by 100 for the optimizer
# (this is a numerical convenience that DOES NOT change t-stats / p-values, only β scale).
y_scale = 100.0
y_garch_input = y_p_logit * y_scale
with warnings.catch_warnings():
    warnings.simplefilter("ignore")
    garch_model = arch_model(
        y_garch_input,
        x=X_no_const,            # 3 lagged-X regressors as mean-equation exogenous covariates
        mean="LS",                # least-squares (constant + exogenous) mean
        vol="Garch",
        p=1,
        q=1,
        dist="normal",
        rescale=False,
    )
    garch_fit = garch_model.fit(disp="off", show_warning=False)

# Extract X-lag mean-equation coefficients. arch indexes them as 'x0', 'x1', 'x2' for the
# three exogenous regressors (X_lag6, X_lag9, X_lag12 in input order). Convert back from
# the y_scale=100 rescaling for reporting (β_garch_composite scaled to logit-Y units).
garch_params = garch_fit.params
garch_pvals_two = garch_fit.pvalues
garch_tstats = garch_fit.tvalues
print(f"  GARCH(1,1)-X param idx: {list(garch_params.index)}")

x_keys = [k for k in garch_params.index if k.startswith("x") or k in ("X_lag6", "X_lag9", "X_lag12")]
# arch consistently labels exogenous regressors x0, x1, x2 when passed as numpy array
expected_keys = ["x0", "x1", "x2"]
for k in expected_keys:
    assert k in garch_params.index, f"GARCH param {k} not found in {list(garch_params.index)}"

beta_garch_lag6_raw = float(garch_params["x0"])
beta_garch_lag9_raw = float(garch_params["x1"])
beta_garch_lag12_raw = float(garch_params["x2"])
# rescale back from y_scale=100 to logit-Y units
beta_garch_lag6 = beta_garch_lag6_raw / y_scale
beta_garch_lag9 = beta_garch_lag9_raw / y_scale
beta_garch_lag12 = beta_garch_lag12_raw / y_scale
beta_garch_composite = beta_garch_lag6 + beta_garch_lag9 + beta_garch_lag12

# Composite SE via covariance of the three coefficients (unaffected by y_scale rescaling
# in the t-stat sense — but for completeness reconstruct in logit-Y units):
cov_xlag = garch_fit.params.index.get_indexer(["x0", "x1", "x2"])
cov_full = garch_fit.param_cov.values  # full param covariance
cov_3 = cov_full[np.ix_(cov_xlag, cov_xlag)] / (y_scale ** 2)  # rescale to logit-Y units
c_garch = np.array([1.0, 1.0, 1.0])
var_composite = float(c_garch @ cov_3 @ c_garch)
se_garch_composite = float(np.sqrt(var_composite)) if var_composite > 0 else float("nan")
t_garch_composite = beta_garch_composite / se_garch_composite if se_garch_composite > 0 else float("nan")

# One-sided p (positive-direction alternative per §3.4 D-ii): p_one = 1 - Φ(t)
p_one_garch_composite = float(1.0 - stats.norm.cdf(t_garch_composite)) if not np.isnan(t_garch_composite) else float("nan")

print(f"  β_garch_lag6           : {beta_garch_lag6:+.8f}")
print(f"  β_garch_lag9           : {beta_garch_lag9:+.8f}")
print(f"  β_garch_lag12          : {beta_garch_lag12:+.8f}")
print(f"  β_garch_composite      : {beta_garch_composite:+.8f}")
print(f"  SE_garch_composite     : {se_garch_composite:.8f}")
print(f"  t-stat_garch_composite : {t_garch_composite:+.6f}")
print(f"  p_one_garch_composite  : {p_one_garch_composite:.8e}")

sign_garch_composite = 1 if beta_garch_composite > 0 else -1 if beta_garch_composite < 0 else 0
d_ii_sign_match = (sign_garch_composite == sign_primary) and (sign_garch_composite > 0)
d_ii_pass = d_ii_sign_match and (p_one_garch_composite <= 0.10)
print(f"  D-ii sign match prim   : {d_ii_sign_match} (sign_primary={sign_primary:+d}, sign_garch={sign_garch_composite:+d})")
print(f"  D-ii α=0.10 threshold  : {'PASS' if d_ii_pass else 'FAIL'} (β>0 AND p_one≤0.10)")

# ── §7.5. D-iii — EVT POT (spec §5.5 D-iii + §3.4 D-iii) ──────────────────
# Per spec §3.4 D-iii verbatim: "in peaks-over-threshold (POT) on upper-tail residuals
# from the primary OLS, the regression of exceedances on log(COP_USD_{t-9}) yields a
# positive coefficient at one-sided p ≤ 0.10".
# Per spec §5.5 D-iii verbatim: "peaks-over-threshold on upper-tail residuals from the
# primary OLS (threshold = empirical 0.90 residual quantile); exceedances regressed on
# log(COP_USD_{t-9}). Tests whether extreme positive residual realizations covary with X
# — the most direct EVT-style test of tail-region β."
print(f"\n=== D-iii — EVT POT (peaks-over-threshold) ===")
# Step 1: extract primary OLS residuals
e_primary = result_primary.resid
# Step 2: empirical 0.90 residual quantile threshold
u_threshold = float(np.percentile(e_primary, 90))
# Step 3: identify upper-tail exceedances (residuals strictly above threshold)
exceedance_mask = e_primary > u_threshold
e_exceedances = e_primary[exceedance_mask] - u_threshold
n_exceedances = int(exceedance_mask.sum())
print(f"  Primary residual range : [{e_primary.min():+.6f}, {e_primary.max():+.6f}]")
print(f"  Threshold u (0.90 q)   : {u_threshold:+.8f}")
print(f"  N exceedances          : {n_exceedances} (of {len(e_primary)} = {n_exceedances/len(e_primary):.1%})")

# Step 4: GPD diagnostic fit (informational — confirms upper-tail is well-described by GPD
# parametric family; spec §3.4 D-iii does not gate on GPD shape parameter ξ but reports it).
gpd_xi, gpd_loc, gpd_scale = genpareto.fit(e_exceedances, floc=0.0)
print(f"  GPD ξ (shape, info)    : {gpd_xi:+.6f}")
print(f"  GPD scale (info)       : {gpd_scale:+.6f}")

# Step 5: regress exceedances on log(COP_USD_{t-9}) — spec §3.4 D-iii anchor on lag-9.
# X_lag9 in this panel IS log(COP_USD_{t-9}) per env.py panel-construction convention.
x_lag9_exc = panel.loc[exceedance_mask, "X_lag9"].to_numpy(dtype=float)
X_pot = sm.add_constant(x_lag9_exc, has_constant="add")
result_pot = sm.OLS(e_exceedances, X_pot).fit(cov_type="HC3")  # heteroskedasticity-robust SE
beta_pot = float(result_pot.params[1])
se_pot = float(result_pot.bse[1])
t_pot = beta_pot / se_pot if se_pot > 0 else float("nan")
p_one_pot = float(1.0 - stats.norm.cdf(t_pot)) if not np.isnan(t_pot) else float("nan")
print(f"  β_pot (exc ~ X_lag9)   : {beta_pot:+.8f}")
print(f"  SE_pot                 : {se_pot:.8f}")
print(f"  t-stat_pot             : {t_pot:+.6f}")
print(f"  p_one_pot              : {p_one_pot:.8e}")

sign_pot = 1 if beta_pot > 0 else -1 if beta_pot < 0 else 0
d_iii_sign_match = (sign_pot == sign_primary) and (sign_pot > 0)
d_iii_pass = d_iii_sign_match and (p_one_pot <= 0.10)
print(f"  D-iii sign match prim  : {d_iii_sign_match} (sign_primary={sign_primary:+d}, sign_pot={sign_pot:+d})")
print(f"  D-iii α=0.10 threshold : {'PASS' if d_iii_pass else 'FAIL'} (β>0 AND p_one≤0.10)")

# ── §7.6. §3.4 disjunction evaluation (PASS via D-i / D-ii / D-iii / NONE) ──
print(f"\n=== §3.4 ESCALATE-PASS disjunction evaluation ===")
disjunct_passes = {"D-i_quantile_tau_0.90": d_i_pass,
                   "D-ii_garch_x_composite": d_ii_pass,
                   "D-iii_evt_pot": d_iii_pass}
n_disjuncts_pass = sum(int(v) for v in disjunct_passes.values())
escalate_pass = any(disjunct_passes.values())
escalate_pass_via = [k for k, v in disjunct_passes.items() if v] or ["NONE"]
verdict_label = "ESCALATE-PASS" if escalate_pass else "ESCALATE-FAIL"
print(f"  D-i quantile τ=0.90    : {'PASS' if d_i_pass else 'FAIL'}")
print(f"  D-ii GARCH-X composite : {'PASS' if d_ii_pass else 'FAIL'}")
print(f"  D-iii EVT POT          : {'PASS' if d_iii_pass else 'FAIL'}")
print(f"  Disjuncts passing      : {n_disjuncts_pass} of 3")
print(f"  Disjunction verdict    : {verdict_label}")
print(f"  ESCALATE-PASS via      : {escalate_pass_via}")

# ── §7.7. ESCALATION_RESULTS.md emission ──────────────────────────────────
OUT_PATH = OUTPUT_DIR / "ESCALATION_RESULTS.md"

NOW = _dt.datetime.now(_dt.timezone.utc).isoformat(timespec="seconds")
SPEC_SHA = "7c72292516f58f3cf2d16464d4f84c3e7d7270ad2c5d3d8ed8fef6b3b2751f5a"  # spec v1.0.2 decision_hash
PLAN_TAG = "v1.1.1"

md_lines = [
    "---",
    f"emitted_at_utc: {NOW}",
    f"emitted_from: notebooks/03_tests_and_sensitivity.ipynb (Trio 6 — §5.5 escalation suite)",
    f"spec_decision_hash: {SPEC_SHA}",
    f"plan_tag: {PLAN_TAG}",
    f"panel_n: {N_EXPECTED_POST_LAG_12}",
    f"window: {panel['year_month'].min().strftime('%Y-%m')} → {panel['year_month'].max().strftime('%Y-%m')}",
    "anti_fishing_framing: pre-pinned convex-payoff evidence test (per spec §9.6); NOT post-hoc rescue",
    "---",
    "",
    "# ESCALATION_RESULTS — Trio 6 §5.5 escalation suite (D-i / D-ii / D-iii)",
    "",
    "Spec v1.0.2 §5.5 escalation methodology + §3.4 ESCALATE-PASS disjunction. Per spec §9.6 the suite is "
    "PRE-AUTHORIZED co-primary convex-payoff evidence; framing is **pre-pinned convex-payoff evidence test, "
    "ran whether or not mean-OLS passed** — anti-fishing-banned to call this a rescue claim. Pair D Stage-1 "
    "precedent (`project_pair_d_phase2_pass`): same suite pre-authorized at the same numerics; Pair D PASSed "
    "mean-OLS and did not need to invoke the suite, but the inheritance is verbatim.",
    "",
    f"**Sign anchor (re-fit primary OLS):** `β_composite_primary = {beta_composite_primary:+.8f}` "
    f"(sign = {'+' if sign_primary > 0 else '−' if sign_primary < 0 else '0'}).",
    "",
    "## §1 — D-i: quantile regression τ=0.90 (spec §5.5 D-i + §3.4 D-i)",
    "",
    "**Specification (verbatim from spec §5.5 D-i + §3.4 D-i).** Conditional quantile regression of "
    "`Y_t (logit)` on the three lagged-X regressors (`X_lag6`, `X_lag9`, `X_lag12`) plus intercept, "
    "estimated at `τ = 0.90`. Per §3.4 D-i the threshold trigger evaluates the lag-9 coefficient "
    "`β_qr_lag9 (τ=0.90)` (spec-pinned 'representative middle-of-window lag') against `> 0` AND "
    "one-sided `p ≤ 0.10`.",
    "",
    "**Numerics.**",
    f"- `β_qr_lag9 (τ=0.90)` = `{beta_qr_lag9:+.8f}`",
    f"- `SE_qr_lag9` = `{se_qr_lag9:.8f}`",
    f"- `t-stat_qr_lag9` = `{t_qr_lag9:+.6f}`",
    f"- `p_one_qr_lag9` = `{p_one_qr_lag9:.8e}`",
    f"- `β_qr_composite (info)` = `{beta_qr_composite:+.8f}` (sum of three lag coefficients; for cross-reference)",
    f"- sign match primary: `{d_i_sign_match}` (`sign_primary` = {sign_primary:+d}, `sign_qr_lag9` = {sign_qr_lag9:+d})",
    "",
    f"**§3.4 D-i α = 0.10 threshold evaluation:** `{'PASS' if d_i_pass else 'FAIL'}` "
    f"(requires `β > 0` AND `p_one ≤ 0.10`).",
    "",
    "## §2 — D-ii: GARCH(1,1)-X (spec §5.5 D-ii + §3.4 D-ii)",
    "",
    "**Specification (verbatim from spec §5.5 D-ii + §3.4 D-ii).** GARCH(1,1) on `Y_t (logit)` with the "
    "three lagged-X regressors (`X_lag6`, `X_lag9`, `X_lag12`) entering the mean equation as exogenous "
    "covariates. Composite GARCH-X β = sum of three lag mean-equation coefficients. Per §3.4 D-ii the "
    "threshold trigger evaluates the composite β against `> 0` AND one-sided `p ≤ 0.10`.",
    "",
    "**Numerics.**",
    f"- `β_garch_lag6` = `{beta_garch_lag6:+.8f}` (logit-Y units, post-rescaling from y_scale=100 optimizer convenience)",
    f"- `β_garch_lag9` = `{beta_garch_lag9:+.8f}`",
    f"- `β_garch_lag12` = `{beta_garch_lag12:+.8f}`",
    f"- `β_garch_composite` = `{beta_garch_composite:+.8f}`",
    f"- `SE_garch_composite` = `{se_garch_composite:.8f}` (linear-restriction SE via param covariance)",
    f"- `t-stat_garch_composite` = `{t_garch_composite:+.6f}`",
    f"- `p_one_garch_composite` = `{p_one_garch_composite:.8e}`",
    f"- sign match primary: `{d_ii_sign_match}` (`sign_primary` = {sign_primary:+d}, `sign_garch` = {sign_garch_composite:+d})",
    "",
    f"**§3.4 D-ii α = 0.10 threshold evaluation:** `{'PASS' if d_ii_pass else 'FAIL'}` "
    f"(requires `β > 0` AND `p_one ≤ 0.10`).",
    "",
    "## §3 — D-iii: EVT POT (spec §5.5 D-iii + §3.4 D-iii)",
    "",
    "**Specification (verbatim from spec §5.5 D-iii + §3.4 D-iii).** Peaks-over-threshold on upper-tail "
    "residuals from the primary OLS at empirical 0.90 residual quantile threshold; exceedances regressed "
    "on `log(COP_USD_{t-9})` (= `X_lag9` per env.py panel-construction convention). Per §3.4 D-iii the "
    "threshold trigger evaluates the regression coefficient against `> 0` AND one-sided `p ≤ 0.10`.",
    "",
    "**Numerics.**",
    f"- Primary residual range: `[{e_primary.min():+.6f}, {e_primary.max():+.6f}]`",
    f"- Threshold u (0.90 quantile): `{u_threshold:+.8f}`",
    f"- N exceedances: `{n_exceedances}` of `{len(e_primary)}` ({n_exceedances/len(e_primary):.1%})",
    f"- GPD shape ξ (informational): `{gpd_xi:+.6f}` (positive ⇒ heavy-tail; near-zero ⇒ exponential-tail; negative ⇒ bounded-tail)",
    f"- GPD scale (informational): `{gpd_scale:+.6f}`",
    f"- `β_pot (exceedance ~ X_lag9)` = `{beta_pot:+.8f}`",
    f"- `SE_pot` (HC3-robust) = `{se_pot:.8f}`",
    f"- `t-stat_pot` = `{t_pot:+.6f}`",
    f"- `p_one_pot` = `{p_one_pot:.8e}`",
    f"- sign match primary: `{d_iii_sign_match}` (`sign_primary` = {sign_primary:+d}, `sign_pot` = {sign_pot:+d})",
    "",
    f"**§3.4 D-iii α = 0.10 threshold evaluation:** `{'PASS' if d_iii_pass else 'FAIL'}` "
    f"(requires `β > 0` AND `p_one ≤ 0.10`).",
    "",
    "## §4 — §3.4 ESCALATE-PASS disjunction status",
    "",
    "Per spec §3.4 verbatim: *ESCALATE-PASS fires if any one or more of the three disjuncts hold; "
    "otherwise ESCALATE-FAIL.* Per §3.4 structural-disjunction MHT defense: each disjunct estimates a "
    "*distinct distributional-moment parameter* (D-i upper-tail conditional response; D-ii conditional-mean "
    "under time-varying-volatility; D-iii extreme-residual covariation) mapping to a *distinct convex-instrument "
    "design family* (D-i range-LP / covered-call; D-ii volatility-conditional perpetual; D-iii tail-risk "
    "OTM put); MHT-correction does NOT apply since this is structural-disjunction over distinct parameters, "
    "not multiple identical tests on the same parameter.",
    "",
    "**Per-disjunct verdicts:**",
    f"- D-i quantile τ=0.90: **{'PASS' if d_i_pass else 'FAIL'}**",
    f"- D-ii GARCH(1,1)-X composite: **{'PASS' if d_ii_pass else 'FAIL'}**",
    f"- D-iii EVT POT: **{'PASS' if d_iii_pass else 'FAIL'}**",
    "",
    f"**Disjuncts passing:** `{n_disjuncts_pass}` of 3.",
    "",
    f"**§3.4 disjunction verdict:** **{verdict_label}** "
    f"({'via ' + ', '.join(escalate_pass_via) if escalate_pass else 'no disjunct fires'}).",
    "",
    "## §5 — §9.6 pre-authorization framing acknowledgment",
    "",
    "Per spec §9.6 verbatim: *Escalation as pre-authorization, not post-hoc rescue. The §5.5 + §3.4 "
    "escalation suite was pre-authorized in this spec before any data was pulled. Framing escalation in "
    "the result memo as 'rescue' is anti-fishing-banned; the framing must be 'pre-pinned convex-payoff "
    "evidence test, ran whether or not mean-OLS passed'.* This Trio 6 emission was authored regardless of "
    "Trio 5's routing-branch outcome (PASS-with-κ-clear / PASS-with-κ-not-clear / ESCALATE / FAIL); the "
    "escalation-suite execution is governed by §9.6 pre-authorization, not by Trio 5's per-row sign-AGREE "
    "tally. Pair D Stage-1 precedent (`project_pair_d_phase2_pass`): the same suite was pre-authorized at "
    "the same numerics and Pair D's PASS verdict at §3.1 meant Pair D did not invoke the suite — but the "
    "*pre-authorization* discipline is verbatim and cross-iteration-binding.",
    "",
    "## §6 — Phase-3 result memo + gate_verdict.json integration directive",
    "",
    "Per spec §9.17 invariant (Phase-3 result-memo §11.X disclosure pre-pin), the Phase-3 result memo "
    "MUST cite this Trio 6 escalation-suite output verbatim in §11.X(b) (3-row R1/R3 + escalation-suite "
    "adjudication table) and §11.X(d) (Stage-2 hedge-geometry mapping conditional on which disjunct(s) "
    "fire). Per the dispatch contract, NB02 Trio 3 final-status emission of `output/gate_verdict.json` "
    "MUST add a top-level field `escalation_results` containing: per-disjunct β + SE + t + p_one + pass-flag; "
    "disjunction-verdict label; ESCALATE-PASS-via list; spec-decision-hash + plan-tag pin. The integration "
    "directive is binding for Phase 2 closure prior to Phase 2.5 (Economist Analyst framework application).",
    "",
    "**End of ESCALATION_RESULTS.**",
    "",
]

OUT_PATH.parent.mkdir(parents=True, exist_ok=True)
OUT_PATH.write_text("\n".join(md_lines), encoding="utf-8")
print(f"\n=== ESCALATION_RESULTS.md emitted ===")
print(f"  path: {OUT_PATH}")
print(f"  size: {OUT_PATH.stat().st_size} bytes")

# ── §7.8. Trio 6 readout summary ──────────────────────────────────────────
print(f"\n=== Trio 6 readout summary ===")
print(f"  D-i quantile τ=0.90    : {'PASS' if d_i_pass else 'FAIL'}  (β_qr_lag9={beta_qr_lag9:+.6e}, p_one={p_one_qr_lag9:.4e})")
print(f"  D-ii GARCH(1,1)-X      : {'PASS' if d_ii_pass else 'FAIL'}  (β_composite={beta_garch_composite:+.6e}, p_one={p_one_garch_composite:.4e})")
print(f"  D-iii EVT POT          : {'PASS' if d_iii_pass else 'FAIL'}  (β_pot={beta_pot:+.6e}, p_one={p_one_pot:.4e})")
print(f"  §3.4 disjunction       : {verdict_label} ({n_disjuncts_pass}/3 disjuncts pass)")
print(f"  ESCALATE-PASS via      : {escalate_pass_via}")
print(f"  §9.6 framing           : pre-pinned convex-payoff evidence test (NOT post-hoc rescue)")
print(f"  Pair D contrast        : Pair D pre-authorized same suite at same numerics (memory `project_pair_d_phase2_pass`); did not invoke at PASS verdict")
print(f"  NB03 trio status       : 6 of 6 trios complete; NB03 closed")
print(f"  Phase 2 closure        : unblocked — proceed to Phase 2.5 Economist Analyst framework application")
```
